```python
import pandas as pd

data_path = "/tmp/youtube_trending_videos_global.csv"

df = pd.read_csv(data_path)
print(df.shape)
df.head()
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)
df.head()
```

```python
needed_cols = [
    "video_id",
    "video_title",
    "video_description",
    "video_tags",
    "video_category_id",
    "video_published_at",
    "video_trending__date",
    "video_view_count",
    "video_like_count",
    "video_comment_count",
    "video_trending_country",
]

df = df[needed_cols].copy()
print("After selecting needed cols:", df.shape)
```

```python
num_cols_raw = ["video_view_count", "video_like_count", "video_comment_count"]

for c in num_cols_raw:
    df[c] = pd.to_numeric(
        df[c].astype(str).str.replace(",", "", regex=False),
        errors="coerce"
    )

df = df.dropna(subset=num_cols_raw)
print("After numeric cleanup:", df.shape)
```

```python
df["video_title"] = df["video_title"].astype(str)
df["video_tags"] = df["video_tags"].fillna("").astype(str)
df["video_description"] = df["video_description"].fillna("").astype(str)
```

```python
df["title_len"] = df["video_title"].str.len()
df["desc_len"]  = df["video_description"].str.len()
df["tags_count"] = df["video_tags"].apply(
    lambda x: 0 if x in ("[none]", "") else len(str(x).split("|"))
)

eps = 1e-6
df["likes_per_view"]    = df["video_like_count"] / (df["video_view_count"] + eps)
df["comments_per_view"] = df["video_comment_count"] / (df["video_view_count"] + eps)

df["video_published_at"] = pd.to_datetime(df["video_published_at"], errors="coerce")
df["publish_hour"] = df["video_published_at"].dt.hour
df["publish_dow"]  = df["video_published_at"].dt.dayofweek
```

```python
df = df.dropna(subset=["video_trending_country", "video_trending__date"])

def make_viral_flag(s):
    q = s.quantile(0.90)
    return (s >= q).astype(int)

df["viral"] = (
    df.groupby(["video_trending_country", "video_trending__date"])["video_view_count"]
      .transform(make_viral_flag)
)
```

```python
df = df[df["viral"].notna()].copy()
df["viral"] = df["viral"].astype(int)

print("After viral label creation:", df.shape)
print(df["viral"].value_counts(normalize=True))
```

```python
import numpy as np

numeric_cols = [
    "video_view_count",
    "video_like_count",
    "video_comment_count",
    "title_len",
    "desc_len",
    "tags_count",
    "likes_per_view",
    "comments_per_view",
    "publish_hour",
    "publish_dow"
]

categorical_cols = [
    "video_category_id",
    "video_trending_country"
]

for c in categorical_cols:
    df[c] = df[c].astype(str)

X_num = df[numeric_cols].fillna(0)
df_cat = pd.get_dummies(df[categorical_cols], drop_first=True)

X_meta = np.hstack([X_num.values, df_cat.values])
y = df["viral"].astype(int).values

print("X_meta shape:", X_meta.shape)
print("Any NaN in y?", np.isnan(y).any())
```

```python
from sklearn.model_selection import train_test_split

X_train_meta, X_temp_meta, y_train, y_temp = train_test_split(
    X_meta, y,
    test_size=0.2, random_state=42, stratify=y
)

X_val_meta, X_test_meta, y_val, y_test = train_test_split(
    X_temp_meta, y_temp,
    test_size=0.5, random_state=42, stratify=y_temp
)
```

Scale metadata

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_meta_scaled = scaler.fit_transform(X_train_meta)
X_val_meta_scaled   = scaler.transform(X_val_meta)
X_test_meta_scaled  = scaler.transform(X_test_meta)

meta_input_dim = X_train_meta_scaled.shape[1]
meta_input_dim
```

```python
# Build the full text column
df["text_input"] = (
    df["video_title"] + " " +
    df["video_tags"] + " " +
    df["video_description"]
)

text_data = df["text_input"].values

# Now split text along with metadata
(
    X_train_meta, X_temp_meta,
    y_train, y_temp,
    train_text, temp_text
) = train_test_split(
    X_meta, y, text_data,
    test_size=0.2, random_state=42, stratify=y
)

(
    X_val_meta, X_test_meta,
    y_val, y_test,
    val_text, test_text
) = train_test_split(
    X_temp_meta, y_temp, temp_text,
    test_size=0.5, random_state=42, stratify=y_temp
)

print("Text train examples:", len(train_text))
print("Text val examples:", len(val_text))
print("Text test examples:", len(test_text))
```

Tokenization

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

MAX_VOCAB = 50000
MAX_LEN = 200

tokenizer = Tokenizer(num_words=MAX_VOCAB, oov_token="<OOV>")
tokenizer.fit_on_texts(train_text)

X_train_text = tokenizer.texts_to_sequences(train_text)
X_val_text   = tokenizer.texts_to_sequences(val_text)
X_test_text  = tokenizer.texts_to_sequences(test_text)

X_train_text = pad_sequences(X_train_text, maxlen=MAX_LEN, padding="post")
X_val_text   = pad_sequences(X_val_text, maxlen=MAX_LEN, padding="post")
X_test_text  = pad_sequences(X_test_text, maxlen=MAX_LEN, padding="post")
```

MLP MODEL

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

def build_mlp(input_dim):
    inp = keras.Input(shape=(input_dim,))
    x = layers.Dense(256, activation="relu")(inp)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.4)(x)
    x = layers.Dense(128, activation="relu")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)
    x = layers.Dense(64, activation="relu")(x)
    out = layers.Dense(1, activation="sigmoid")(x)

    model = keras.Model(inp, out)
    model.compile(
        optimizer=keras.optimizers.Adam(1e-3),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return model

mlp_model = build_mlp(meta_input_dim)
mlp_model.summary()

history_mlp = mlp_model.fit(
    X_train_meta_scaled, y_train,
    validation_data=(X_val_meta_scaled, y_val),
    epochs=20,
    batch_size=1024,
    callbacks=[keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)]
)
```

LSTM

```python
def build_lstm():
    inp = keras.Input(shape=(MAX_LEN,))
    x = layers.Embedding(MAX_VOCAB, 128)(inp)
    x = layers.LSTM(128, return_sequences=False)(x)
    x = layers.Dropout(0.4)(x)
    x = layers.Dense(64, activation="relu")(x)
    out = layers.Dense(1, activation="sigmoid")(x)

    model = keras.Model(inp, out)
    model.compile(
        optimizer=keras.optimizers.Adam(1e-3),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return model

lstm_model = build_lstm()
lstm_model.summary()

history_lstm = lstm_model.fit(
    X_train_text, y_train,
    validation_data=(X_val_text, y_val),
    epochs=5,
    batch_size=1024,
    callbacks=[keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)]
)
```

CNN

```python
def build_cnn():
    inp = keras.Input(shape=(MAX_LEN,))
    x = layers.Embedding(MAX_VOCAB, 128)(inp)
    x = layers.Conv1D(128, 5, activation="relu")(x)
    x = layers.MaxPooling1D(2)(x)
    x = layers.Conv1D(128, 5, activation="relu")(x)
    x = layers.GlobalMaxPooling1D()(x)
    x = layers.Dropout(0.4)(x)
    x = layers.Dense(64, activation="relu")(x)
    out = layers.Dense(1, activation="sigmoid")(x)

    model = keras.Model(inp, out)
    model.compile(
        optimizer=keras.optimizers.Adam(1e-3),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return model

cnn_model = build_cnn()
cnn_model.summary()

history_cnn = cnn_model.fit(
    X_train_text, y_train,
    validation_data=(X_val_text, y_val),
    epochs=5,
    batch_size=1024,
    callbacks=[keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)]
)
```

FUSION MODEL

```python
def build_fusion(meta_dim):
    # Metadata branch
    meta_in = keras.Input(shape=(meta_dim,))
    m = layers.Dense(128, activation="relu")(meta_in)
    m = layers.BatchNormalization()(m)
    m = layers.Dropout(0.3)(m)

    # Text branch
    text_in = keras.Input(shape=(MAX_LEN,))
    t = layers.Embedding(MAX_VOCAB, 128)(text_in)
    t = layers.Conv1D(128, 5, activation="relu")(t)
    t = layers.MaxPooling1D(2)(t)
    t = layers.LSTM(128)(t)
    t = layers.Dropout(0.3)(t)

    # Combine
    combined = layers.concatenate([m, t])
    x = layers.Dense(128, activation="relu")(combined)
    x = layers.Dropout(0.3)(x)
    out = layers.Dense(1, activation="sigmoid")(x)

    model = keras.Model([meta_in, text_in], out)
    model.compile(
        optimizer=keras.optimizers.Adam(1e-3),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return model

fusion_model = build_fusion(meta_input_dim)
fusion_model.summary()

history_fusion = fusion_model.fit(
    [X_train_meta_scaled, X_train_text],
    y_train,
    validation_data=([X_val_meta_scaled, X_val_text], y_val),
    epochs=7,
    batch_size=1024,
    callbacks=[keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)]
)
```

Evaluate models

```python
print("MLP Test:")
mlp_model.evaluate(X_test_meta_scaled, y_test)

print("LSTM Test:")
lstm_model.evaluate(X_test_text, y_test)

print("CNN Test:")
cnn_model.evaluate(X_test_text, y_test)

print("Fusion Test:")
fusion_model.evaluate([X_test_meta_scaled, X_test_text], y_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import pandas as pd
import numpy as np

def evaluate_model(model, X_meta=None, X_text=None):
    if X_meta is not None and X_text is not None:
        preds = model.predict([X_meta, X_text]).ravel()
    elif X_meta is not None:
        preds = model.predict(X_meta).ravel()
    else:
        preds = model.predict(X_text).ravel()

    preds_bin = (preds > 0.5).astype(int)

    return {
        "Accuracy": accuracy_score(y_test, preds_bin),
        "AUC": roc_auc_score(y_test, preds),
        "Precision": precision_score(y_test, preds_bin, zero_division=0),
        "Recall": recall_score(y_test, preds_bin, zero_division=0),
        "F1 score": f1_score(y_test, preds_bin, zero_division=0)
    }

results = {}

results["MLP"] = evaluate_model(mlp_model, X_meta=X_test_meta_scaled)
results["LSTM"] = evaluate_model(lstm_model, X_text=X_test_text)
results["CNN"] = evaluate_model(cnn_model, X_text=X_test_text)
results["Fusion"] = evaluate_model(fusion_model,
                                   X_meta=X_test_meta_scaled,
                                   X_text=X_test_text)

summary_df = pd.DataFrame(results).T
summary_df
```
